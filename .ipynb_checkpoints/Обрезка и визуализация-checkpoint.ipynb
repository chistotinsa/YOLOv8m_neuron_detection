```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import random

# Размер пикселя в микрометрах - оригинал
pixel_width_um = 0.3450  

# Разрешение JPEG 4096 x 3008
image_width_px = 2048  # Ширина изображения в пикселях //Заменить!

# Вычисление соотношения пикселей к миллиметрам
pixel_to_mm_ratio = (pixel_width_um / 1000) / image_width_px

annotation_file = "C:\\Users\\minik\\Downloads\\Neuron.v4i.yolov8_640_Ruslan\\train\\labels\\snaps-10-_jpg.rf.5a1e7d69ddd88ff283724496a48fa54e.txt"

image_path = "C:\\Users\\minik\\Downloads\\Neuron.v4i.yolov8_640_Ruslan\\train\\images\\snaps-10-_jpg.rf.5a1e7d69ddd88ff283724496a48fa54e.jpg"
```

```python
# Функция для преобразования пикселей в миллиметры
def pixel_to_mm(x_pixel, y_pixel):
    x_mm = x_pixel * pixel_to_mm_ratio
    y_mm = y_pixel * pixel_to_mm_ratio
    return x_mm, y_mm
```

```python
def parse_yolo_annotation(annotation_file, image_path):
    annotations = []
    with open(annotation_file, 'r') as f:
        for line in f:
            data = line.strip().split()
            class_id = {0:'neuron'}
            image = cv2.imread(image_path)
            image_height, image_width, _ = image.shape
            x_center, y_center, width, height =map(float, data[1:])
            # Перевод относительных координат в абсолютные пиксельные координаты
            x_center_px = int(x_center * image_width)
            y_center_px = int(y_center * image_height)
            width_px = int(width * image_width)
            height_px = int(height * image_height)
    
    # Вычисление координат bbox
            x1 = int(x_center_px - width_px / 2)
            y1 = int(y_center_px - height_px / 2)
            x2 = int(x_center_px + width_px / 2)
            y2 = int(y_center_px + height_px / 2)
            annotations.append((x1, y1, x2, y2))
            
    return annotations
```

```python
list_annotations = parse_yolo_annotation(annotation_file, image_path)
```

```python
def parse_yolo_annotation(yolo_pred, image_path):
    image = cv2.imread(image_path)
    annotations = []
    for data in yolo_pred:
        print(data)
        class_id = {0:'neuron'}
        image_height, image_width, _ = image.shape
        x_center, y_center, width, height =map(float, data)
        # Перевод относительных координат в абсолютные пиксельные координаты
        x_center_px = int(x_center * image_width)
        y_center_px = int(y_center * image_height)
        width_px = int(width * image_width)
        height_px = int(height * image_height)
    
        # Вычисление координат bbox
        x1 = int(x_center_px - width_px / 2)
        y1 = int(y_center_px - height_px / 2)
        x2 = int(x_center_px + width_px / 2)
        y2 = int(y_center_px + height_px / 2)
        annotations.append((x1, y1, x2, y2))
            
    return annotations

parse_yolo_annotation(list_annotations, image_path)
```

```python
# Получить координаты X и Y для всех аннотаций
x_coordinates = [annotation[0] for annotation in list_annotations]
y_coordinates = [annotation[1] for annotation in list_annotations]

# Вычислить среднее значение и стандартное отклонение для координат X и Y
mean_x = np.mean(x_coordinates)
std_x = np.std(x_coordinates)
mean_y = np.mean(y_coordinates)
std_y = np.std(y_coordinates)

# Вычислить Z-оценки для координат X и Y для каждой аннотации
z_scores_x = [(x - mean_x) / std_x for x in x_coordinates]
z_scores_y = [(y - mean_y) / std_y for y in y_coordinates]

# Определить порог Z для исключения выбросов
threshold_z = 1.5

# Исключить аннотации с выбросами
annotations = []
for i in range(len(list_annotations)):
    if abs(z_scores_x[i]) <= threshold_z and abs(z_scores_y[i]) <= threshold_z:
        annotations.append(list_annotations[i])
```

```python
# Загрузка изображения
image = cv2.imread(image_path)

# Размеры участка для вырезания 
# В jpg оригинале 4095х3008 пикселей, посчитав на оригинальном 
# CYZ снимке количество пикселей на микрометры, пришел к выводу, что 300 мкм (это ~ 870 пикселей)
# Поэтому 870px в нашем изображении будет ~ 300um

patch_size = 161  #//Заменить!
half_patch_size = patch_size // 2 # от рандомной точки на линии тренда откладываем половину размера в 4 стороны


# Извлечение координат центров прямоугольников
x_values = [(rect[0] + rect[2]) / 2 for rect in annotations]
y_values = [(rect[1] + rect[3]) / 2 for rect in annotations]

# Выполнение линейной регрессии
p = np.polyfit(x_values, y_values, 4)

# Создание изогнутой линии тренда
x_trend_pixels = np.linspace(min(x_values), max(x_values), 100)
y_trend_pixels = np.polyval(p, x_trend_pixels)

# Минимальное расстояние от края изображения до выбираемой точки
min_distance_from_edge = patch_size / 2 

while True:
    # Если точка не удовлетворяет условиям, выбираем новый случайный индекс
    random_index = random.randint(0, len(x_trend_pixels) - 1)
    # Получение координат x и y для случайного индекса на линии тренда
    random_point_x = x_trend_pixels[random_index]
    random_point_y = y_trend_pixels[random_index]
    
# Проверка, находится ли точка на достаточном расстоянии от краев изображения
    if (random_point_x >= min_distance_from_edge and random_point_x <= image.shape[0] - min_distance_from_edge and
        random_point_y >= min_distance_from_edge and random_point_y <= image.shape[1] - min_distance_from_edge):
        # Если да, выходим из цикла
        break
    
    # Если точка не удовлетворяет условиям, выбираем новый случайный индекс
    random_index = random.randint(0, len(x_trend_pixels) - 1)

# Обработать края фотографии - рандом +-150 мкм!


print(random_point_x, random_point_y)
# Вычисление границ участка изображения
top_left_x = int(random_point_x - half_patch_size)
top_left_y = int(random_point_y - half_patch_size)
bottom_right_x = int(random_point_x + half_patch_size)
bottom_right_y = int(random_point_y + half_patch_size)

# Обрезка участка изображения
patch = image[top_left_y:bottom_right_y, top_left_x:bottom_right_x]

# Подсчет количества аннотаций в области обрезания
num_annotations_in_patch = 0
for annotation in annotations:
    x1, y1, x2, y2 = annotation
    # Проверка, попадает ли прямоугольник в область обрезания
    if x1 >= top_left_x and y1 >= top_left_y and x2 <= bottom_right_x and y2 <= bottom_right_y:
        num_annotations_in_patch += 1

print("Количество аннотаций в области интереса:", num_annotations_in_patch)

annotated_patch = patch.copy()

# Рисуем прямоугольники аннотаций на обрезанном участке изображения
for annotation in annotations:
    x1, y1, x2, y2 = annotation
    # Проверяем, попадает ли аннотация в область обрезания
    if x1 >= top_left_x and y1 >= top_left_y and x2 <= bottom_right_x and y2 <= bottom_right_y:
        # Рисуем прямоугольник
        cv2.rectangle(annotated_patch, (x1 - top_left_x, y1 - top_left_y), (x2 - top_left_x, y2 - top_left_y), (0, 255, 0), 2)
        cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)

# Вывод изображений
# Создаем третий рисунок с линией тренда и аннотациями
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Оригинальное изображение
axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[0].set_title('Оригинальное изображение')
axes[0].axis('off')

# Визуализация линни тренда
axes[1].plot(x_trend_pixels, y_trend_pixels, color='red', linestyle='--', label='Линия тренда')
axes[1].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[1].scatter(x_values, y_values, label='Прямоугольники')
axes[1].set_title('Линия тренда из прямоугольников')
axes[1].grid(True)


# Обрезанный участок с аннотациями
axes[2].imshow(cv2.cvtColor(annotated_patch, cv2.COLOR_BGR2RGB))
axes[2].set_title('Обрезанный участок с аннотациями')
axes[2].axis('off')



plt.show()
```
